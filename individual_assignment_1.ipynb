3. In a group project, the main branch is shared by every member, so it acts as the main working branch. It is useful to work in a separate branch (away from main) so that individual members can test code without affecting everyone else. Otherwise, if someone pushes experimental / nonworking code into main, it affects everyone else.

   The branch-pull-request workflow also keeps main clean, and not flooded with multiple small commits from every team member.
